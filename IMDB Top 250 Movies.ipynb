## RUNTIME 


### The average run time of movies in every IMDB year

```python
# average run time
avg_runtime = imdB.groupby('IMDByear')['RunTime'].agg(np.mean) 
avg_runtime
```

```python
# the average run time trend by year
plt.figure(figsize = (12, 5))
avg_runtime.plot()
```

There is an increase in the average runtime of movies as the year passess by.

```python
# null values
imdB.isna().sum()
```

### Movies with run time > 200mins(3 hours) 

```python
# Movies with RunTime >= 200
Top_by_RunTime = imdB[imdB['RunTime'] >= 200].sort_values('RunTime', ascending = False) 
Top_by_RunTime.drop_duplicates(subset='Title')
```

### The movie with the Maximum runtime in each IMDB year

```python
yearly_max_runtime = imdB.loc[imdB.groupby('IMDByear')['RunTime'].idxmax()] 
yearly_max_runtime
```

```python
# max run time graphical display 
plt.figure(figsize = (10, 4))
imdB.groupby('IMDByear')['RunTime'].max().plot()
```

The max run time of eacg IMDB year is fairly averagely the same until there was an increase in 2014

## SCORE 


#### Movies that score 90 points and above

```python
# Movies that score 90 points and above 
top_by_score = imdB[imdB['Score'] >= 90]
top_by_score.drop_duplicates(subset='Title').sort_values('Score', ascending = False)
```

### Movie with the Maximum Score in each IMDB year

```python
year_score_max = imdB.loc[imdB.groupby('IMDByear').Score.idxmax()] 
year_score_max
```

# RATING 


### Top rated movies

```python
# Top rated movies 
top_rated = imdB[imdB['Rating'] > 8.5] 
top_rated.drop_duplicates(subset='Title').sort_values('Rating', ascending = False)
```

### The Topmost rated movie in each IMDB year

```python
year_rate_max = imdB.loc[imdB.groupby('IMDByear').Rating.idxmax()] 
year_rate_max
```

The Shawshank Redemption has appeared as the topmost rated movie in every of the IMDB year so far.

## GROSS 



### Movies that grossed the most earnings

```python
# Top gross earners
imdB.sort_values('Gross', ascending=False).head(10)
```

```python
# Movies with high rating that Gross High 
gross_rating = imdB[(imdB['Gross'] > 500) & (imdB['Rating'] >= 7.0)]
gross_rating.drop_duplicates(subset='Title').sort_values(['Gross', 'Rating'], False, False)
```

### Average Gross earnings in IMDB year

```python
yearly_gross_avg = imdB.groupby('IMDByear')['Gross'].mean() 
yearly_gross_avg
```

```python
# Average gross earnings trend over IMDB year
plt.figure(figsize = (12, 5))
yearly_gross_avg.plot()
```

##### There is an increase in the average gross earnings that appeared in the IMDBTop250
